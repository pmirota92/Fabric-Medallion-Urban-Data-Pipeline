```python
from pyspark.sql.functions import count, max

# Odczytujemy "czyste" dane z Silver
silver_df = spark.table("silver_bus_current")

# Agregacja pod wykres
gold_df = silver_df.groupBy("line").agg(
    count("bus_id").alias("active_units"),
    max("signal_time").alias("last_update")
)

# Zapis do Gold (zawsze nadpisujemy aktualny stan)
gold_df.write.format("delta").mode("overwrite").saveAsTable("gold_line_summary")

print("Warstwa GOLD gotowa! Dane są przygotowane do Power BI.")
display(gold_df)
```
